```python
#Apply scaling techniques to numerical features:
# a.Standard Scaling
num_col = df.select_dtypes(include=['int64','float64']).columns
num_col
```

```python
from sklearn.preprocessing import StandardScaler
std_sacler = StandardScaler()
df_standard_scaler = df.copy()
df_standard_scaler[num_col] = std_sacler.fit_transform(df[num_col])
```

```python
df_standard_scaler[num_col]
```

```python
#b. Min-Max Scaling
from sklearn.preprocessing import MinMaxScaler
minmax_scale = MinMaxScaler()
df_minmax_scale = df.copy()
df_minmax_scale[num_col] = minmax_scale.fit_transform(df[num_col])
```

```python
df_minmax_scale[num_col]
```

```python
#Discuss the scenarios where each scaling technique is preferred and why.
#:- StandardScaler we use When data follows a normal distribution. because it centers the data around 0 with unit variance.
#:- MinMaxScaler we use When data does not follow a normal distribution. because It preserves the original data distribution.
```

```python
#Task2: Encoding Techniques:
#Apply One-Hot Encoding to categorical variables with less than 5 categories.
cat_col = df.select_dtypes(include=['object']).columns
cat_col
```

```python
one_hot_col = [col for col in cat_col if df[col].nunique() <= 5]
one_hot_col
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop='first', sparse_output=False)
df_one_hot = encoder.fit_transform(df[one_hot_col])
df_one_hot = pd.DataFrame(df_one_hot, columns=encoder.get_feature_names_out(one_hot_col))
```

```python
df_one_hot
```

```python
#Use Label Encoding for categorical variables
from sklearn.preprocessing import LabelEncoder
label_enc = LabelEncoder()
df_label_encoder = df.copy()

for col in cat_col:
    df_label_encoder[col] = label_enc.fit_transform(df_label_encoder[col])
```

```python
df_label_encoder[cat_col]
```

```python
#Load the dataset and conduct basic data exploration (summary statistics, missing values, data types).
df_final = pd.concat([df_standard_scaler[num_col], df_label_encoder[cat_col]], axis=1)

df_final
```

```python
#Task3: Feature Engineering:
#Create at least 2 new features that could be beneficial for the model. Explain the rationale behind your choices.
# Create new feature: Work Hours Category
def work_hours_category(hours):
    if hours < 30:
        return 'Part_Time'
    elif hours <= 40:
        return 'Full_Time'
    else:
        return 'Over_Time'

df['WorkHoursCategory'] = df['hours_per_week'].apply(work_hours_category)

# Create new feature: Age Group
def age_group(age):
    if age < 30:
        return 'Young'
    elif age < 50:
        return 'Middle_Aged'
    else:
        return 'Senior'

df['AgeGroup'] = df['age'].apply(age_group)
```
