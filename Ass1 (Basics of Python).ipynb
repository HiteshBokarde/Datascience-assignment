```python
#Exercise 1: Prime Numbers

num = int(input("Enter the number: "))
if num > 1:
  for i in range(2 , int(num ** 0.5) + 1):
    if (num % i) == 0:
      print(num, "is not a prime number")
      break
  else:
      print(num, "is a prime number")
else:
  print(num, "is not a prime number")
```

```python
#Exercise 2: Product of Random Numbers

import random

num1 = random.randint(1, 10)
num2 = random.randint(1,10)

print("What is the product of", num1, "*", num2)
user_ans = int(input("Your Answer: "))

if user_ans == num1*num2:
  print("Your answer is correct")
else:
  print("Your answer is wrong, The Correct answer is", num1*num2)
```

```python
#Exercise 3: Squares of Even/Odd Numbers

#Choice: Even numbers

for num in range(100, 201):
  if num % 2 ==0:
    print("The Square of ", num, "is", num**2)
```

```python
#Exercise 3: Squares of Even/Odd Numbers

#Choice: odd numbers

for num in range(100, 201):
  if num % 2 != 0:
    print("The Square of ", num, "is", num**2)
```

```python
#Exercise 4: Hospital Billing System

room_charge_per_day = 2000
doctor_fee = 1500
lab_test_charge = 300
discount = 0.10

days_stayed = 4
num_lab_tests = 3
medicine_charges = 2400

room_total = room_charge_per_day * days_stayed
lab_total = lab_test_charge * num_lab_tests
total_before_discount = room_total + doctor_fee + lab_total + medicine_charges

if total_before_discount > 10000:
    discount = total_before_discount * discount
else:
    discount = 0

final_bill = total_before_discount - discount

print("------ Hospital Billing System ------")
print(f"Room Charges (₹2000 × {days_stayed} days): ₹{room_total}")
print(f"Doctor Consultation Fee: ₹{doctor_fee}")
print(f"Lab Test Charges (₹300 × {num_lab_tests} tests): ₹{lab_total}")
print(f"Medicine Charges: ₹{medicine_charges}")
print("-------------------------------------")
print(f"Total (Before Discount): ₹{total_before_discount}")
print(f"Discount Applied: ₹{discount}")
print("-------------------------------------")
print(f"Final Bill Amount: ₹{final_bill}")
```
