```python
# Sec A: Data Structures & Control Structure
#1.
list= [10,20,30,40,50]

#append:
list.append(60)
print("After Append: ",list)

#extend:
list.extend([70,80])
print("After Extend: ",list)

#insert:
list.insert(3,25)
print("After Insert: ",list)

#remove:
list.remove(50)
print("After Remove: ",list)

#pop
list.pop(3)
print("After Pop: ",list)

# index
print("Index of 30: ",list.index(30))

# count
print("Count of 20: ",list.count(20))

# sort
list.sort()
print("After Sort: ",list)

# reverse
list.reverse()
print("After Reverse: ",list)

#clear
list.clear()
print("After Clear: ",list)
```

```python
#Section C: Exploring pandas

data = {
    "Name": ["soun", "monu", "mandar", "pratik", "pinku", "riya", "radha", "uma", "titu", "nitya"],
    "Age": [17, 21, 22, 23, 24, 25, 26, 27, 28, 29],
    "City": ["pune", "amt", "delhi", "nagpur", "chennai",
             "pune", "amt", "delhi", "nagpur", "chennai"],
    "Salary": [50500, 60500, None, 55500, 65500, None, 70500, 62500, 58500, 72500]
}

df = pd.DataFrame(data)
print(df)

#info and describe
print(df.info())

print(df.describe())
```

```python
#select column
print(df[["Name","City"]])
```

```python
#Drop column
df = df.drop(columns=['City'])
print(df)
```

```python
#missing values with mean

df['Salary'].fillna(df['Salary'].mean(), inplace=True)
print(df)
```

```python
#Remove Duplicates
df = df.drop_duplicates()
print(df)
```
