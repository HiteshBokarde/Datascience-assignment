```python
#Hypothesis Testing:

# Null Hypothesis:No evidence that Weekly Operating costs are higher
# Alternate Hypothesis:There is evidence that Weekly Operating costs are higher

import math
from scipy.stats import norm

sample_mean= 3050
population_mean= 1000+ 5*600
std_deviation= 5 * 25
n= 25
alpha= 0.05

z = (sample_mean - population_mean) / (std_deviation / math.sqrt(n))

critical_z = norm.ppf(1 - alpha)

print("Z-Statistic:", z)
print("Critical Z-Value:", critical_z)

if z > critical_z:
    print("Reject the null hypothesis: There is evidence that Weekly Operating costs are higher.")
else:
    print("Fail to reject the null hypothesis: No evidence that Weekly Operating costs are higher.")
```

```python
#Confidence Interval:

import numpy as np
from scipy.stats import t

data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85,
                 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])

n = len(data)
mean = np.mean(data)
std_sample = np.std(data, ddof=1)
alpha = 0.01

# t critical value
t_critical = t.ppf(1 - alpha/2, df=n-1)

# Margin of error
margin_error = t_critical * (std_sample / np.sqrt(n))

# Confidence Interval
ci_lower = mean - margin_error
ci_upper = mean + margin_error

print("Sample Mean:", round(mean, 4))
print("99% Confidence Interval (Using Sample SD):")
print(f"({round(ci_lower, 4)}, {round(ci_upper, 4)})")
```

```python
from scipy.stats import norm

sigma_population = 0.2

# z critical value
z_critical = norm.ppf(1 - alpha/2)

# Margin of error
margin_error_z = z_critical * (sigma_population / np.sqrt(n))

# Confidence Interval
ci_lower_z = mean - margin_error_z
ci_upper_z = mean + margin_error_z

print("99% Confidence Interval (Using Known Population SD):")
print(f"({round(ci_lower_z, 4)}, {round(ci_upper_z, 4)})")
```
